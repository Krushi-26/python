What is a Class?
Concept:  A class is a blueprint or template. It defines what an object should look like and what it can do. You write it once, then create as many objects from it as you need.
Think of it as:  A class is a cookie cutter. An object is the actual cookie made from it. The cutter defines the shape — each cookie is a real, usable thing.
How to Create a Class
Syntax:  class ClassName:  — use the 'class' keyword, then the name in PascalCase, then a colon.
class Dog:           # Define a class called Dog
    pass             # 'pass' means nothing inside yet — avoids IndentationError
If you write 'class Dog:' with nothing below it and Python finds something else — it throws IndentationError. Always put 'pass' if you want an empty class.

```python
class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

    def introduce(self):
        return f"My name is {self.name}, I am {self.age} years old and I study {self.course}."


# Creating objects
student1 = Student("Krushi", 22, "Python")
student2 = Student("nagireddy", 21, "SQL")

# Adding grades
student1.add_grade(90)
student1.add_grade(95)
student1.add_grade(85)

student2.add_grade(70)
student2.add_grade(75)

# Printing details
print(student1.introduce())
print("Average:", student1.calculate_average())

print(student2.introduce())
print("Average:", student2.calculate_average())
   
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, I am {self.name}."


class Student(Person):
    def __init__(self, name, course):
        super().__init__(name)
        self.course = course

    def greet(self):
        return f"{super().greet()} I study {self.course}."


student = Student("Krushi", "Python")
print(student.greet())
```

#### Inheritance
Inheritance = One class gets properties and methods from another class.
Real life:
Student IS-A Person
So Student can inherit from Person.

```python
#A constructor is a special method that runs automatically when an object is created.
# In Python → __init__
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        return f"{self.name} scored {self.marks}"


s1 = Student("Krushi", 90)
print(s1.display())
```

```python
# Single Inheritance
# One parent → one child
class Person:
    def __init__(self, name):
        self.name = name
class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)
        self.marks = marks
s1 = Student("Krushi", 95)
print(s1.name)
print(s1.marks)
# super() calls parent class constructor.
# It means:
# "Go to Person class and run its init"
# So Student doesn’t rewrite name logic.
```

```python
# Multilevel Inheritance
# Grandparent → Parent → Child
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    pass

class Topper(Student):
    def award(self):
        return f"{self.name} gets admitted in class"

t1 = Topper("Krushi")
print(t1.award())
# Topper inherits:
# From Student
# From Person (indirectly)
# So Topper has everything.
```

```python
# Hierarchical Inheritance
# One parent → Multiple children
#Both Student and Teacher inherit from Person.
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    pass

class Teacher(Person):
    pass

s1 = Student("Krushi")
t1 = Teacher("Nagireddy")

print(s1.name)
print(t1.name)
```

```python
# Multiple Inheritance?
# One child → Multiple parents
class Academic:
    def study(self):
        return "Studies hard"

class Sports:
    def play(self):
        return "Plays ludo"

class Student(Academic, Sports):
    pass

s1 = Student()
print(s1.study())
print(s1.play())
```

```python
# Encapsulation means:
# Bundle data + methods together
# Hide internal implementation
# Control access
# Taking a car as example:
# You press start button.
# You don’t manually ignite spark plugs.
# The internal complexity is hidden.
# One class contains another class.
# Car have a engine. we hide engine logic inside the car
class Engine:
    def start(self):
        return "Engine started"
class Car:
    def __init__(self):
        self.engine = Engine()   # Composition
    def start(self):
        return self.engine.start()
car = Car()
print(car.start())
```

```python
# Dynamic extension 
# The behavior is decided at runtime based on input
class TextUpper:
    def format(self, text):
        return text.upper()
class TextLower:
    def format(self, text):
        return text.lower()
class Text:
    def __init__(self, style):
        if style == "upper":
            self.formatter = TextUpper()
        else:
            self.formatter = TextLower()
    def display(self, text):
        return self.formatter.format(text)

t1 = Text("upper")
print(t1.display("hello"))
t2 = Text("lower")
print(t2.display("HELLO"))
```

```python
# Duck Typing
# Python doesn’t check type.
# It checks behavior.
# No inheritance needed.
# Just same method name.
class Dog:
    def speak(self):
        return "Bark"
class Cat:
    def speak(self):
        return "Meow"
def make_sound(animal):
    print(animal.speak())
make_sound(Dog())
make_sound(Cat())
```
