BMI Calculator
The body mass index (BMI) is a measure used in medicine to see if someone is underweight or overweight. This is the formula used to calculate it:
bmi is equal to the person's weight divided by the person's height squared.


```python
height = 1.65 
weight = 84 
bmi = weight / height ** 2
print(bmi)
```

TIP CALCULATOR 

```python
print("tip calculator ")
bill = float(input("total bill? $"))
tip = float(input("what percent tip would you like to give? 10 15 18"))
split = int(input("how many ppl to split the bill"))
tip_as_payment = tip/100
total_tip_amount = bill*tip_as_payment
total_bill = bill + total_tip_amount
bill_per_person = (total_bill /split)
Final_amount = round(bill_per_person,2)
print(f"each person should pay {Final_amount}")
```

### BMI Calculator with Interpretations
Add some if/elif/else statements to the BMI calculator so that it interprets the BMI values calculated.
If the bmi is under 18.5 (not including), print out "underweight"
If the bmi is between 18.5 (including) and 25 (not including), print out "normal weight"
If the bmi is 25 (including) or over, print out "overweight"
Refer to this graphic for help:

```python
weight = 85
height = 1.85

bmi = weight / (height ** 2)

if bmi >= 25:
    print("overweight")
elif bmi >= 18.5:
    print("normal weight")
else:
    print("underweight")
```
