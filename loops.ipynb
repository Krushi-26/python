```python
def count_greater_than_5(numbers):
    count = 0

    for num in numbers:
        if num > 5:
            count += 1

    return count


nums = [2, 7, 4, 9, 1]
print(count_greater_than_5(nums))
```

```python
import random

# Word list
word_list = ["python", "banana", "computer", "hangman", "keyboard"]

# Hangman stages
stages = [
'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''',
'''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''',
'''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''',
'''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''',
'''
  +---+
  |   |
      |
      |
      |
      |
=========
'''
]

# Pick random word
chosen_word = random.choice(word_list)
word_length = len(chosen_word)

# Game variables
lives = 6
end_of_game = False

# Create display with blanks
display = []
for _ in range(word_length):
    display.append("_")

print("Welcome to Hangman!")
print(stages[lives])

# Game loop
while not end_of_game:

    guess = input("Guess a letter: ").lower()

    # Check if already guessed
    if guess in display:
        print("You already guessed that letter.")

    # Check each letter in chosen word
    for position in range(word_length):
        letter = chosen_word[position]
        if letter == guess:
            display[position] = guess

    # If guess not in word
    if guess not in chosen_word:
        lives -= 1
        print(f"Wrong guess. Lives left: {lives}")
        if lives == 0:
            end_of_game = True
            print("You lose!")
            print(f"The word was: {chosen_word}")

    print(" ".join(display))
    print(stages[lives])

    # Check win
    if "_" not in display:
        end_of_game = True
        print("You win!")
```
